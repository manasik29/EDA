```python
df = pd.read_excel('/content/Country-Code.xlsx')
df.head()
```

```python
final_df = pd.merge(data, df, on='Country Code', how='left')
final_df.head(2)
```

```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
plt.pie(final_df.Country.value_counts()[:3].values, labels = country_names[:3], shadow=True, radius=2, explode=(0.5, 0,0), autopct="%1.2f%%")
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette=['pink','red','orange','yellow','green','green'])
```
