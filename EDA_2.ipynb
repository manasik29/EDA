# **ISOLATION FOREST**

```python
from sklearn.ensemble import IsolationForest
```

```python
data2 = pd.read_csv('/content/iris.csv', index_col=0)
```

```python
data2=pd.get_dummies(data2)
```

```python
data2
```

```python
#training the model
clf = IsolationForest(random_state=10, contamination=.015)
clf.fit(data2)
```

```python
y_outliers = clf.predict(data2)
```

```python
y_outliers
```

```python
data2.loc[149] = [20,30,50,40,1,0,0]
```

```python
clf.fit(data2)
y_outliers=clf.predict(data2)
y_outliers
```

```python
data2['Scores'] = clf.decision_function(data2)
```

```python
data2['anamoly'] = clf.predict(data2.iloc[:,0:7])
```
