```python
import pandas as pd
df = pd.read_excel('/content/questions.xlsx')
df.head()
```

```python
#create a new feature Sale by multiplying product_quantity and price_per_item
Sale = df['product_quantity']*df['price_per_item']
Sale
```

```python
#add Sale column to our df
df.insert(4, 'Sale', Sale)
df
```
